## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=8):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    imshape = img.shape   # helps to extrapolate also in challenge video   
    top_border_line_drawing = 50
    
    # extrapolate funciton that draws lines from bottom to top
    
    def calc_m(y2, y1, x2, x1):   # calculate slope m 
        if (x2-x1) == 0:
            return 0
        return (y2 - y1) / (x2 - x1)
    
    def calc_b(y2, y1, x2, x1):   # calculate y-intersect b
        if (x2-x1) == 0:
            return y1
        m = (y2 - y1) / (x2 - x1)
        return y1 - (m*x1) 
    
    def extrapolate(b, m, xtop=None, color=[255,0,0], top_border_line_drawing=50):
        
        #precaution 0 division
        if m == 0:
            pass
        
        else:
            if xtop != None:  #if lines were about to cross we prevent intersecting of left and right line by adjusting xtop and ytop
                #getting top x,y pair
                xtop = int(xtop)  #redundant but hey
                ytop = int(m*xtop + b)
                ybottom = int(imshape[0])  #boundary bottom of polygon..
                xbottom = int((ybottom - b) / m)   
                cv2.line(img,(xbottom, ybottom), (xtop, ytop), color, thickness) 

            else:
                #getting top x,y pair
                #top y is boundary height of polygon of interest > taken from vertices and adjusted for eye pleasing viewing
                ytop = int((imshape[0]/2 + top_border_line_drawing))
                xtop = int((ytop - b) / m)
                ybottom = int(imshape[0])  #boundary bottom of polygon..
                xbottom = int((ybottom - b) / m)
                # slope from top left to bottom right > lanes right border >> therefore top first, bottom second
                cv2.line(img,(xbottom, ybottom), (xtop, ytop), color, thickness) 
    
    
    # getting averages
    
    rightborder_x1_array = []
    rightborder_y1_array = []
    rightborder_x2_array = []
    rightborder_y2_array = []
    leftborder_x1_array = []
    leftborder_y1_array = []
    leftborder_x2_array = []
    leftborder_y2_array = []
         
    #collecting points and sorting them to average them out later
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            
            #precaution 0 division
            if (x2 - x1) == 0:
                continue
            
            m = (y2 - y1) / (x2 - x1)
            
            if m == 0:
                continue
            
            if 2/5 < m < 4/5:  #right border of lane
                rightborder_x1_array.append(x1) 
                rightborder_y1_array.append(y1) 
                rightborder_x2_array.append(x2)  
                rightborder_y2_array.append(y2)  
                
            elif -1.2 < m < - 0.6:  #left border of lane
                leftborder_x1_array.append(x1) 
                leftborder_y1_array.append(y1) 
                leftborder_x2_array.append(x2)  
                leftborder_y2_array.append(y2)
                
            else:
                continue   # continue without noise (=unwanted lines
    
    #taking the average
    def avg(L):
        ttl = 0
        length = len(L)
        if length == 0:
            length = 1
        for i in L:
            ttl += i
        return (ttl / length)
    
    rbx1 = avg(rightborder_x1_array)
    rby1 = avg(rightborder_y1_array)
    rbx2 = avg(rightborder_x2_array)
    rby2 = avg(rightborder_y2_array)
    lbx1 = avg(leftborder_x1_array)
    lby1 = avg(leftborder_y1_array)
    lbx2 = avg(leftborder_x2_array)
    lby2 = avg(leftborder_y2_array)
    
    
    m_right = calc_m(rby2, rby1, rbx2, rbx1)
    b_right = calc_b(rby2, rby1, rbx2, rbx1)
    m_left = calc_m(lby2, lby1, lbx2, lbx1)
    b_left = calc_b(lby2, lby1, lbx2, lbx1)
    
    # for challenge video: check if extrapolated lines intersect 
    def crossing_lines(xtop_left,xtop_right):
        # if lines cross: extrapolate using extra parameter
        delta_top_xs = xtop_left - xtop_right
        xtop_right = int(xtop_right + (delta_top_xs/2) + 6)  # the magic number in the end is to keep them from actually touching
        xtop_left = int(xtop_left - (delta_top_xs/2) - 6)
        extrapolate(b_right, m_right, xtop=xtop_right, color=[0,100,255], top_border_line_drawing=50)
        extrapolate(b_left, m_left, xtop=xtop_left, color=[0,255,100], top_border_line_drawing=50)
     
    '''right border of lane'''
    if 2/5 < m_right < 4/5:    
        # prepare for check if lines cross
        ytop_right = int((imshape[0]/2 + top_border_line_drawing))
        xtop_right = int((ytop_right - b_right) / m_right)
        if m_left != 0:    # preventing 0 division error
            ytop_left = int((imshape[0]/2 + top_border_line_drawing))
            xtop_left = int((ytop_left - b_left) / m_left)
            if xtop_left > xtop_right: # check if lines cross: extrapolate using extra parameter
                crossing_lines(xtop_left,xtop_right)
            else:  # if its an image without crossing lines: extrapolate:
                extrapolate(b_right, m_right, xtop=None, color=[0,0,255], top_border_line_drawing=50)
        else:  # better than to loose lines, so extrapolate: 
            extrapolate(b_right, m_right, xtop=None, color=[0,0,255], top_border_line_drawing=50)
                


    '''left border of lane'''        
    if -1.2 < m_left < - 0.6:  
        ytop_left = int((imshape[0]/2 + top_border_line_drawing))
        xtop_left = int((ytop_left - b_left) / m_left)
        if m_right != 0:  # preventing 0 division error
            ytop_right = int((imshape[0]/2 + top_border_line_drawing))
            xtop_right = int((ytop_right - b_right) / m_right)
            if xtop_left > xtop_right: # check if lines cross: extrapolate using extra parameter
                crossing_lines(xtop_left,xtop_right)
            else:  # if its an image without crossing lines: extrapolate:                
                extrapolate(b_left, m_left, xtop=None, color=[255,0,0], top_border_line_drawing=50)
        else:  # better than to loose lines, so extrapolate:                
            extrapolate(b_left, m_left, xtop=None, color=[255,0,0], top_border_line_drawing=50)



def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

````python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)

    #pic_ = mpimg.imread("test_images/"+pic)
    gray = grayscale(image)

    kernel_size = 5
    blurred_gray = gaussian_blur(gray, kernel_size)

    low_threshold = 50
    high_threshold = 150
    edgy = canny(blurred_gray, low_threshold, high_threshold)

    #using helper function to mask image
    imshape = image.shape    #to be used in vertices
    vertices = np.array([[(0,imshape[0]),(imshape[1]/2 -15, imshape[0]/2 + 45),  #+45 to have it low enough for the horizon..
                          (imshape[1]/2 + 15, imshape[0]/2 + 45),
                          (imshape[1],imshape[0])]],dtype=np.int32)

    #vertices to be used for region of interest:

    masked_edges = region_of_interest(edgy, vertices)

    rho = 1 #2 is good with theta /180 # tresh 50 # minline 3 # maxlinegap 150
    theta = np.pi/180 #/100 might keep it out of the cars in front -- in combi with rho1,thresh 50, minline 20, maxgap 3 #orig 180
    threshold = 50
    min_line_len = 3
    max_line_gap = 150

    lines = hough_lines(masked_edges, rho, theta, threshold, min_line_len, max_line_gap)

    initial_img = image

    result = weighted_img(lines, initial_img, α=0.8, β=1., γ=0.)

    return result
````
